# Dealing With Outliers

```python
# Show trends in numerical columns
df_movie.plot(subplots=True,figsize=(16,35))
plt.show()
```

```python
per25th=df_movie["budget"].quantile(0.25)
print(per25th)
per75th=df_movie["budget"].quantile(0.75)
print(per75th)
```

```python
IQR=per75th-per25th
IQR
```

```python
upper= per75th+(IQR*1.5)
lower = per25th - ( IQR * 1.5)

print(upper,lower)
```

```python
df_movie[(df_movie["budget"] > upper) | (df_movie["budget"] < lower)].info()
```

```python
df_movie = df_movie[~((df_movie["budget"] > upper) | (df_movie["budget"] < lower))]
```

```python
per25th=df_movie["gross"].quantile(0.25)
print(per25th)
per75th=df_movie["gross"].quantile(0.75)
print(per75th)
```

```python
IQR=per75th-per25th
IQR
```

```python
upper= per75th+(IQR*1.5)
lower = per25th - ( IQR * 1.5)

print(upper,lower)
```

```python
df_movie = df_movie[~((df_movie["gross"] > upper) | (df_movie["gross"] < lower))]
```
